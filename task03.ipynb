```python
# ==========================================
# Task 3: Web Scraper for News Headlines
# Python Developer Internship
# ==========================================

# Install required packages (Run once if needed)
# !pip install requests beautifulsoup4 pandas

import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime

# ------------------------------------------
# Function to fetch headlines
# ------------------------------------------

def fetch_headlines():
    url = "https://www.bbc.com/news"

    headers = {
        "User-Agent": "Mozilla/5.0"
    }

    try:
        response = requests.get(url, headers=headers, timeout=10)
        response.raise_for_status()

        soup = BeautifulSoup(response.text, "html.parser")

        headlines = []

        # Collect headlines from h2 tags
        for tag in soup.find_all("h2"):
            text = tag.get_text(strip=True)

            if text and text not in headlines:
                headlines.append(text)

        return headlines

    except requests.exceptions.RequestException as e:
        print("Error:", e)
        return []


# ------------------------------------------
# Save headlines to TXT file
# ------------------------------------------

def save_to_txt(headlines):
    with open("news_headlines.txt", "w", encoding="utf-8") as file:
        file.write("TOP NEWS HEADLINES\n")
        file.write("=" * 50 + "\n\n")

        for i, headline in enumerate(headlines, start=1):
            file.write(f"{i}. {headline}\n")

    print("Headlines saved to news_headlines.txt")


# ------------------------------------------
# Save headlines to CSV (Extra Feature)
# ------------------------------------------

def save_to_csv(headlines):
    df = pd.DataFrame({
        "Headline Number": range(1, len(headlines) + 1),
        "Headline": headlines,
        "Scraped Time": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    })

    df.to_csv("news_headlines.csv", index=False)
    print("Headlines saved to news_headlines.csv")


# ------------------------------------------
# Display headlines
# ------------------------------------------

def display_headlines(headlines):
    print("\nTOP NEWS HEADLINES")
    print("=" * 50)

    for i, headline in enumerate(headlines, start=1):
        print(f"{i}. {headline}")


# ------------------------------------------
# Main Program
# ------------------------------------------

headlines = fetch_headlines()

if headlines:
    display_headlines(headlines)
    save_to_txt(headlines)
    save_to_csv(headlines)

    print(f"\nTotal Headlines Scraped: {len(headlines)}")

else:
    print("No headlines found.")
```
